```python
class CompteBancaire:
    
    def __init__(self,nom,argent):
        self.nom = nom
        self.solde = argent

    def affiche(self): 
        print('Le solde du compte bancaire de '+self.nom + ' est de : ' + str(self.solde)+' euros')

    def depot(self,depot):
        self.solde = self.solde + depot

    def retrait(self,montant):
        self.solde = self.solde - montant

maxence = CompteBancaire('Maxence', 300)
maxence.affiche()
maxence.depot(1588)
maxence.affiche()
maxence.retrait(388)
maxence.affiche()
```

```python
class Point:
    x = ""
    y = ""
    z = ""

    def __init__(self,x="",y=""):
        self.x = x
        self.y = y

    def __init__(self,x= "",y= "",z = ""):
        self.x = x
        self.y = y
        self.z = z

    def ToString(self): 
        if(self.z ==""):
            print('p('+str(self.x) + ', ' + str(self.y)+')')
        else:
            print('p('+str(self.x) + ', ' + str(self.y)+ ', '+ str(self.z)+')')

P1 = Point(2,3)
P1.ToString()
P2=Point(1,-5,6)
P2.ToString()
```

```python
class DateNaissance:
    
    def __init__(self,jour,mois,annee):
        self.jour = jour
        self.mois = mois
        self.annee = annee

    def ToString(self): 
        return 'Date de naissance : ' + str(self.jour) + '/' + str(self.mois) + '/' + str(self.annee)

class personne:
    nom = ''
    prenom = ''
    dateNaissance = ''
    
    def __init__(self,nom,prenom,date):
        self.nom = nom
        self.prenom = prenom
        self.dateNaissance = date
    
    def afficher(self):
        print('nom: ' + self.nom
             + '\n Prénom: ' + self.prenom
             + '\n Date de naissance: ' + self.dateNaissance.ToString())

class Employe(personne):
    
    salaire = 0
    
    def __init__(self,nom,prenom,date,salaire):
        self.nom = nom
        self.prenom = prenom
        self.dateNaissance = date
        self.salaire = salaire
        
    def afficher(self):
        print('nom: ' + self.nom
             + '\n Prénom: ' + self.prenom
             + '\n Date de naissance: ' + self.dateNaissance.ToString()
             + '\n Salaire: ' + str(self.salaire))

class Chef(Employe):

    service = ''
    
    def __init__(self,nom,prenom,date,salaire,service):
        self.nom = nom
        self.prenom = prenom
        self.dateNaissance = date
        self.salaire = salaire
        self.service = service
    
    def afficher(self):
        print('nom: '+self.nom
             + '\n Prénom: ' + self.prenom
             + '\n Date de naissance: ' + self.dateNaissance.ToString()
             + '\n Salaire: ' + str(self.salaire)
             + '\n service: ' + str(self.service))


P=personne('Ilyass','Math',DateNaissance(1,7,1982))
P.afficher()
print()
E=Employe('Ilyass','Math',DateNaissance(1,7,1985),7865.548)
E.afficher()
print()

Ch=Chef('Ilyass','Math',DateNaissance(1,7,1988),7865.548,'Ressource humaine')
Ch.afficher()
```

```python
class Lettre:
    adressDest = ''
    adrExped  = ''
    poids = ''
    mode = ''
    format = ''
    prixTimbre = ''
    tarifBase = 0
    
    def __init__(self,ad='',ae='',p='',m='',f=''):
        self.adressDest = ad
        self.adrExped = ae
        self.poids = p
        self.mode = m
        self.format = f        
        self.calculTimbre()

    
    def calculTimbre(self):
        
        if(self.format == 'A3'):
            self.tarifBase = 3.50
        elif (self.format == 'A4'):
            self.tarifBase = 2.50
            
        self.prixTimbre = self.tarifBase + 1.0 * self.poids
        
        if(self.mode == 'expresse'):
            self.prixTimbre = self.prixTimbre*2


    def ToString(self):
        print('lettre: \n'
              + 'Adresse destination: ' + self.adressDest
              + '\n Adresse expedition: ' + self.adrExped
              + '\n Poids: ' + str(self.poids)
              + '\n Mode : ' + self.mode
             + '\n format ' + self.format
             + '\n prix du timbre : ' + str(self.prixTimbre))
        
class Colis (Lettre):
    
    volume = 0
    
    def __init__(self,ad='',ae='',p='',m='',v=''):
        self.adressDest = ad
        self.adrExped = ae
        self.poids = p
        self.mode = m
        self.volume = v
        self.calculTimbre()
        
    def calculTimbre(self):
        self.prixTimbre = 0.25 * self.volume + self.poids * 1.0
 
        if(self.mode == 'expresse'):
            self.prixTimbre = self.prixTimbre*2
    
    def ToString(self):
         print('Colis: \n'
              + 'Adresse destination: ' + self.adressDest
              + '\n Adresse expedition: ' + self.adrExped
              + '\n Poids: ' + str(self.poids)
              + '\n Mode : ' + self.mode
              + '\n Volume ' + str(self.volume)
              + '\n prix du timbre : ' + str(self.prixTimbre))
    
        
lettreUne = Lettre('Lille','Paris',80,'normal','A4')
lettreUne.ToString()
C1=Colis('Marrakeche','Barcelone',3500,'expresse',2.25)
C1.ToString()
```
